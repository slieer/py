```python
import seaborn as sns
import matplotlib.pyplot as plt

# 示例数据
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [1000, 1200, 1100, 1300, 1500, 1400]

# 创建线图
sns.lineplot(x=months, y=sales)

# 添加标签和标题
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 生成 x 值
x = np.linspace(-50, 50, 1000)  # 在区间[-5, 5]内生成100个均匀分布的点

# 计算对应的 y 值
y = x**2 + 2*x - 3

# 使用 Seaborn 创建曲线图
sns.lineplot(x=x, y=y, color="blue")

# 添加标签和标题
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y = x^2 + 2x - 3")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 示例数据
products = ["Product A", "Product B", "Product C", "Product D"]
sales = [120, 210, 150, 180]

# 创建柱状图
sns.barplot(x=products, y=sales)

# 添加标签和标题
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales by Category")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 示例数据
scores = [75, 80, 65, 90, 70, 85, 95, 60, 78, 88]

# 创建直方图和密度图
sns.histplot(scores, bins=5, kde=True, color="blue")  # 设置bins表示箱子数量，kde=True表示绘制密度曲线

# 添加标签和标题
plt.xlabel("Scores")
plt.ylabel("Frequency / Density")
plt.title("Score Distribution with Kernel Density Plot")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# 生成随机数据（标准正态分布）
np.random.seed(42)  # 设置随机种子以确保可重复性
mean = 0  # 正态分布的均值
std_dev = 1  # 正态分布的标准差
sample_size = 10000  # 样本数量

data = np.random.normal(mean, std_dev, sample_size)

# 使用Seaborn创建标准正态分布的直方图
sns.histplot(data, kde=True, color="skyblue")

# 添加标签和标题
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Standard Normal Distribution")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime

# 生成示例时间序列数据
np.random.seed(42)
dates = pd.date_range(start="2020-01-01", periods=24, freq="M")
sales = np.random.randint(1000, 5000, size=24)
data = pd.DataFrame({"Date": dates, "Sales": sales})

# 使用Seaborn创建时间序列折线图
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
sns.lineplot(x="Date", y="Sales", data=data, marker="o")

# 添加标题和标签
plt.title("Monthly Sales Time Series")
plt.xlabel("Date")
plt.ylabel("Sales")

# 旋转X轴标签以避免重叠
plt.xticks(rotation=45)

# 显示图表
plt.tight_layout()
plt.show()
```
